```python
df_merged = df_merged.drop(['id'], axis = 1)
```

```python
df['day'] = df['date'].dt.day
df['week'] = df['date'].dt.isocalendar().week
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['week_day'] = df['date'].dt.day_name()
```

####   Anomalies deduction :

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25%)
    Q3 = data[column].quantile(0.75)  # Third quartile (75%)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
```

```python
sales_anomalies = detect_outliers_iqr(df, 'sales')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

### Baseline Model (NaÃ¯ve Forecasting)

```python
data = data.sort_values(by='date')

# Apply NaÃ¯ve Forecasting (shift previous sales value)
data['NaÃ¯ve_Forecast'] = data['sales'].shift(1)

# Drop the first row since it has NaN forecast
data = data.dropna()


mae = mean_absolute_error(data['sales'], data['NaÃ¯ve_Forecast'])
rmse = np.sqrt(mean_squared_error(data['sales'], data['NaÃ¯ve_Forecast']))


print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
# Plot Actual vs. NaÃ¯ve Forecast
plt.figure(figsize=(10,5))
plt.plot(data['date'], data['sales'], label="Actual Sales")
plt.plot(data['date'], data['NaÃ¯ve_Forecast'], linestyle='dashed', label="NaÃ¯ve Forecast", color='red')
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("NaÃ¯ve Forecasting: Actual vs Forecasted Sales")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = data.fillna(0)
```

```python
data = data.set_index('date')
data
```

```python
X = data[['MA_7','MA_30','STD_7','STD_30']]
y = data['sales']
```

```python
# Train-Test Split (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
# Evaluate Model Performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
# Plot Actual vs Predicted Sales
plt.figure(figsize=(10,5))
plt.plot(y_test.index, y_test, label="Actual Sales", marker='o')
plt.plot(y_test.index, y_pred, label="Random Forest Prediction", linestyle="dashed", color="red")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Random Forest Forecasting")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import auto_arima
```

```python
# Ensure data is sorted by Date
data = data.sort_index()

# Visualize the Sales Data
plt.figure(figsize=(10,5))
plt.plot(data.index, data['sales'], label="Sales Data")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Data Over Time")
plt.legend()
plt.show()
```

```python
# Check Stationarity using ADF Test
def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    if result[1] <= 0.05:
        print("Data is stationary")
    else:
        print("Data is NOT stationary (Needs Differencing)")

adf_test(data['sales'])
```

```python
# Differencing (if needed)
data['Sales_diff'] = data['sales'].diff().dropna()

# Plot ACF & PACF to determine p and q
fig, ax = plt.subplots(1, 2, figsize=(12,5))
plot_acf(data['Sales_diff'].dropna(), ax=ax[0])
plot_pacf(data['Sales_diff'].dropna(), ax=ax[1])
plt.show()

# Auto ARIMA to find best (p, d, q)
auto_arima_model = auto_arima(data['sales'], seasonal=False, trace=True)
print(auto_arima_model.summary())

# Fit ARIMA model with best (p, d, q) values
p, d, q = auto_arima_model.order
arima_model = sm.tsa.ARIMA(data['sales'], order=(p, d, q))
arima_result = arima_model.fit()
```

```python
# Forecast future values
forecast_steps = 10  # Adjust as needed
forecast = arima_result.forecast(steps=forecast_steps)
```

```python
# Plot Actual vs Forecast
plt.figure(figsize=(10,5))
plt.plot(df.index, df['Sales'], label="Actual Sales")
plt.plot(pd.date_range(df.index[-1], periods=forecast_steps+1, freq='D')[1:], forecast, label="ARIMA Forecast", linestyle="dashed", color="red")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("ARIMA Forecasting")
plt.legend()
plt.show()

# Calculate MAE and RMSE
mae = mean_absolute_error(df['Sales'][-forecast_steps:], forecast[:forecast_steps])
rmse = np.sqrt(mean_squared_error(df['Sales'][-forecast_steps:], forecast[:forecast_steps]))
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
